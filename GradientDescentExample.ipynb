#### The following function computes the derivative of a function f at point x

```python
def derivative(f,x):
    delta_x = 0.0000000001
    return (f(x+delta_x) - f(x))/delta_x
```

#### The following function raws a function f between points (x_start, y_start) and (x_end,y_end) with tangents at a list of points x_tan 

```python
def draw_function_with_tangents(f,x_tan,x_start,x_end,y_start,y_end,title):
    x = np.linspace(x_start,x_end,100)
    y = f(x) 
    plt.plot(x,y,'b-')
    
    for i in x_tan:
        x_t = i
        y_t = f(x_t)
        y_tan = derivative(f,x_t) * (x - x_t) + y_t 
        plt.plot(x,y_tan,'r-')
    
    x_axis = np.zeros(x.shape[0])
    plt.plot(x,x_axis,'k-')
    plt.axis([x_start,x_end,y_start,y_end])
    plt.xlabel('x-axis')     
    plt.ylabel('y-axis')    
    
    plt.title(title)
    plt.grid()
    plt.show()    
```

Now if you check the above graph - there are two tangent lines - those two lines are drawn at -2 and -2 of the curve.   
If you just check the line,   it is exactly same as the line above,  so the slope is simple - y-change/x-change.

But if you draw these lines at different places like at x=3 or x=4,   the derivative or the slope would be different.
Please note that the derivative is defined at a specific x coordinate not over a range of x.

Example the derivative is meaningful at x=3 but not between x=3 and 4.   The derivative would change within that range.
It can have a value at x=3, 3.1, 3.2 etc.   In otherwords the derivative is defined as the ratio of y-change over x-range
when x-range is as close to zero as possible.

And there is one line at the bottom of the curve,  which is parallel to x-axis.   So no change in y with x.   So the derivative
or the slope is zero.   That is at the minimum point of the curve,   the slope is zero.  Similarly we can show the derivative 
at the maximum is also zero.   We use this condition that slope is zero at minimum or maximum to find out the minimum of the
function.   In this case, we find the minimum of the cost function which looks similar to the above curve.

```python
def derivative(f,x):
    delta_x = 0.000000001
    return (f(x+delta_x) - f(x))/delta_x
```

### The following function draws the function f between (x_start,y_start) and (x_end,y_end) along with its derivatives/tangents

    - no derivative if derivative_num = 0
    - 1st derivative if derivative_num = 1
    - 2nd derivative if derivative_num = 2

```python
def draw_function(f,x_start,x_end,y_start,y_end,x_label,y_label,derivative_num):
    x = np.linspace(x_start,x_end,100)
    y = f(x) 
    plt.plot(x,y,'y-')
    
    if (derivative_num == 1 or derivative_num == 2):
        # compute the 1st derivative
        y_1_der = derivative(f,x)
        #y_2_der = derivative(lambda x:2*(x-2),x)
        
        plt.plot(x,y_1_der,'b-')
        
        if derivative_num == 2:
            # compute the second derivative from the 1st derivative
            y_2_der = np.zeros(len(x))
            for i in range(0,len(x)-1):
                y_2_der[i] = (y_1_der[i+1] - y_1_der[i])/(x[i+1]-x[i])
            y_2_der[len(x)-1] = y_2_der[len(x)-2]
            plt.plot(x,y_2_der,'r-')
    
    x_axis = np.zeros(x.shape[0])
    plt.plot(x,x_axis,'k-')
    plt.axis([x_start,x_end,y_start,y_end])
    plt.xlabel(x_label)     
    plt.ylabel(y_label)    
    plt.grid()
    plt.show()    
```

```python
# draw function alone
draw_function(lambda x: (x-2)**2+0.5,0,4,-3,4,'x-values','function-values',0)
```

```python
# draw function with 1st derivative 
draw_function(lambda x: (x-2)**2+0.5,0,4,-3,4,'x-values','function-values',1)
```

### Points to remember for Gradient Descent

1. The derivative of a function at its minimum is zero
2. The derivative of the function $x^2$ is 2*x 
3. The derivative of a function $(a*x - b)^2$ is 2*(ax-b) * a where the derivative of a*x-b is a
4. The derivative of a constant is zero
5. The derivative of a*x is a
6. The derivative of a function to the right of its minimum is +ve
7. The derivative of a function to the left of its minimum is -ve

#### The following code computes the coordinates corresponding to the minimum of the cost function.   
The coordinates are initialized to a certain value like 0 and then it is updated move opposite to the diretion of the gradient/derivate of the cost function.   The covergence criteria is when the difference in cost is less than a certain value or when the function crosses the minimum point.

```python
m_current = 0
old_cost = 0
for i in range(1,20000):
    # the following is the derivative of the squared error function
    dL = (2/data.shape[0])*sum((m_current*data.rooms-data.price)*data.rooms)
    
    #print('dL ={}'.format(dL))
    # the following is the gradient descent 
    m_new = m_current - 0.001 * dL
    
    #print('m_new ={}'.format(m_new))
    # the following is the cost at the current location
    new_cost = (1/data.shape[0])*sum((m_current*data.rooms-data.price)**2)  
    
    # print the parameters after every 100 iterations
    
    if i%100 == 0:
        print('slope {} derivative {}'.format(m_new,dL))
    
    if old_cost != 0:
        if new_cost > old_cost:
            print("cost has increased from {} to {}".format(old_cost,new_cost))
            break
        
    if np.abs(old_cost - new_cost) < 0.00000001:
        print('old cost {} new cost {} iteration {}'.format(old_cost,new_cost,i))
        break
    old_cost = new_cost
    m_current = m_new
print(m_current)    
```

### Cost or Error Function

The cost or error is the sum of the differences between the predicted and actual values.   We use the Mean Squared Error function where square the differences and average them.  It is also called the Squared Loss Function.  If we have m data points,  the cost 
function would be written as.

It is $L = \dfrac{1}{m}\sum_{i=1}^{m} (\bar{y_i} - y_i)^2$

The Squared Loss function is uniformly differentiable.   This differentiability is the requirement for the Gradient Descent
method for error or cost minimization.

Since our data contains only one input variable ie the number of bedrooms,   we could write our hypothesis function for linear regression as $\bar{y_i} = m*x_i$.   Also we do not have an intercept term because we are assuming our regression line always
passes through the origin. Now our error would be written as

$L = \dfrac{1}{m}\sum_{i=1}^{m} (m * x_i - y_i)^2$

If we differentiate this equation w.r.t m (slope),   we would get 

$\dfrac{\delta{L}}{\delta{m}} = \dfrac{2}{m}\sum_{i=1}^{m} (m * x_i - y_i)*x_i$

Please note that we are differentiating this w.r.t.m not $x_i$ because $x_i$ is the given input data and
We are trying to fit regression lines with different slopes.   




### The following is the Talor series (polynomial) expansion of any function (non-polynomial) for a small value of $\Delta{x}$

$f(x+\Delta{x}) = f(x) + \Delta{x}^Tf'(x) + \dfrac{\Delta{x}^T}{2!} f''(x) \Delta{x} + ........$

Please note that $\Delta{x}$ and f',f'' etc. are  vectors in case of multi-dimensional data.

We will consider the first order (linear) approximation for Gradient Descent, where we need
$\Delta(x)$ to be very small

$f(x+\Delta{x}) = f(x) + \Delta{x}^T \nabla{f} $

If we re-arrange the terms in the above equation, we get

$\nabla{f}  = \dfrac{f(x+\Delta{x}) - f(x)}{\Delta{x}^T} $

This is the fundamental definition of the derivative of the function f w.r.t. x when $\Delta{x}$ is very small


When we are dealing with minimization problems,   we need the consecutive value of

the function to be smaller than the current one.  That is $ f(x+\Delta{x}) < f(x)$

If we substitute $\Delta{x} = -\eta \nabla{f}$ which gives

$x_{i+1} = x_{i} - \eta \nabla{f}$

$f(x_{i+1}-\eta \nabla{f}) = f(x_i) - \eta  (\nabla{f})^T \nabla{f} $

$f(x_{t+1}) = f(x_t) - \eta  (\nabla{f})^T \nabla{f} $

Here $(\nabla{f})^T \nabla{f}$ is the Norm of the vector which is always positive and $\eta$ is also a small +ve quantity.
So the right hand side is always less than $f(x_t)$  indicating $f(x_{t+1}$ is always smaller than $f(x_t)$ as long as
the quantity $\eta \nabla{f}$ is small.

As long as the $\Delta{x}$ and $\eta$ are small and the function is minimized.
